# Function that performs the search and saves the results in a dataframe

```python
#For information about search_type parameter, refer to https://developers.google.com/places/web-service/supported_types
def googleplaces_query_to_pands_dataframe (lng, lat, keyword, radius , max_pages =3):
    places_data_list = []
    query_result = google_places.nearby_search(
    lat_lng={'lat': lat, 'lng': lng}, keyword=keyword,
    radius=radius)
    for place in query_result.places: 
        place.get_details()
        places_data_list.append([place.place_id, place.name, float(place.geo_location['lat']),float(place.geo_location['lng']), place.rating, place.url])
    #default query only returns the first 20
    page =1 
    while (query_result.has_next_page_token and page < max_pages) :
        page +=1 
        query_result_new_page= google_places.nearby_search(pagetoken=query_result.next_page_token)
        for place in query_result_new_page.places: 
            place.get_details()
            places_data_list.append([place.place_id, place.name, float(place.geo_location['lat']),float(place.geo_location['lng']), place.rating, place.url])
    df = pd.DataFrame(places_data_list, columns = ['ID','Name', 'latitude', 'longitude', 'Rating', 'Url'])
    return df
```
